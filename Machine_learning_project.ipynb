IN THIS PROJECT :WE ARE GOING TO WORK ON **ROCK AND MINE** PREDICTION USING SONAR DATA

**ABOUT THIS PROJECT:**
Consider their is a submarine which is going to war.So the opponent country has planted mines under the water and even they are rocks.So we need to build a system which predict whelter object is rock or mine using sonar data which is collected in laboratary.

**Overview:**
In this model we are going to use **LOGISTIC REGRESSION** because its works very well for binary classification.


1.Importing the dependencies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split  #For training and testing data
from sklearn.linear_model import LogisticRegression   
from sklearn.metrics import accuracy_score            #For predicting accuracy
```

2.Data collection and Data processing

```python
import pandas as pd
sonar_df=pd.read_csv("/content/sonar.all-data.csv")
sonar_df.head()
```

```python
#shape:Gives total no of rows and columns in Dataframe
sonar_df.shape
```

```python
#It gives Statistical measures of the data
sonar_df.describe()
```

```python
#it gives total no of  values present in particular columns
sonar_df["Label"].value_counts()
```

M----> Mine

R----> Rock

```python
#Mean of Mine and Rock for each column 
sonar_df.groupby("Label").mean()
```

```python
#Separating data and label
x=sonar_df.drop(columns="Label",axis=1)  #independent columns 
y=sonar_df["Label"]                      #dependent columns
```

```python
print(x)
```

```python
print(y)
```

3.Training and Testing

```python
from sklearn.model_selection import train_test_split  #For training and testing data

train_x,test_x,train_y,test_y = train_test_split(x , y, test_size=0.1 , stratify=y , random_state=1)
```

```python
print(x.shape,train_x.shape,test_x.shape)
```

```python
print(train_x)
print(train_y)
```

4.MODEL TRAINING--->LOGISTIC REGRESSION

```python
#Training the logistic regression with training data

from sklearn.linear_model import LogisticRegression   

model = LogisticRegression()

model.fit(train_x,train_y)
```

5.MODEL EVALUATION

```python
#Accuracy on training data

from sklearn.metrics import accuracy_score 

train_x_prediction = model.predict(train_x)

training_data_accuracy = accuracy_score(train_x_prediction,train_y)
```

```python
print("Accuracy on training data : " , training_data_accuracy)        #TRAINING ACCURACY
```

```python
#Accuracy on testing data

from sklearn.metrics import accuracy_score 

test_x_prediction = model.predict(test_x)

testing_data_accuracy = accuracy_score(test_x_prediction,test_y)
```

```python
print("Accuracy on testing data : " , testing_data_accuracy)           #TESTING ACCURACY
```
